# json 1000

```python
from faker import Faker
import pandas as pd

fake = Faker()

data = []
for _ in range(1000):                              #repeat the next steps 1000 times to generate 1000 fake records.
    data.append({                                  #adds a dictionary with name, email, address to the list.
            "full_name": fake.name(),                                  # String
            "age": random.randint(18, 70),                             # Integer
            "salary": round(random.uniform(30000, 150000), 2),         # Float
            "is_active": random.choice([True, False]),                 # Boolean
            "register_date": fake.date(),                              # Date
            "login_time": fake.time(),                                 # Time
            "timestamp": fake.date_time(),                             # Datetime
            "gender": random.choice(["Male", "Female", "Other"]),      # Category
            "user_id": str(uuid.uuid4()),                              # UUID
            "favorite_numbers": [random.randint(1, 100) for _ in range(3)]  # List (array)
    })

df = pd.DataFrame(data)                                                 #Saves the DataFrame to a JSON file named "data_1000.json".
df.to_json("data_1000.json", orient="records", indent=4)


import time

# Write time
start = time.time()                                                 #records the current time before writing
df.to_json("data_1000.json", orient="records", indent=4)            #writes the DataFrame to JSON.
end = time.time()                                                   # records the time after writing.
print("Write Time JSON:", end - start)                              #calculates how long it took to write the file.

# Read time
start = time.time()                                                 # records time before reading the JSON.
df2 = pd.read_json("data_1000.json")
end = time.time()
print("Read Time JSON:", end - start)

```

```python
# 1. Import Libraries
from faker import Faker
import pandas as pd
import time

# 2. Generate Fake Data (1000 Records)
fake = Faker()
data = []

for _ in range(1000):                                           # repeat the next steps 1000 times to generate 1000 fake records.
    data.append({
            "full_name": fake.name(),                                  # String
            "age": random.randint(18, 70),                             # Integer
            "salary": round(random.uniform(30000, 150000), 2),         # Float
            "is_active": random.choice([True, False]),                 # Boolean
            "register_date": fake.date(),                              # Date
            "login_time": fake.time(),                                 # Time
            "timestamp": fake.date_time(),                             # Datetime
            "gender": random.choice(["Male", "Female", "Other"]),      # Category
            "user_id": str(uuid.uuid4()),                              # UUID
            "favorite_numbers": [random.randint(1, 100) for _ in range(3)]  # List (array)
    })

df = pd.DataFrame(data)                        #Converts the list of dictionaries into a pandas DataFrame (table).

# 3. Save Data in Arrow Format (.arrow file)
start = time.time()
df.to_feather("data_1000.arrow")             #  DATA is saved in arrow format here, Feather is a file format used to store data. It is based on  Arrow and used mainly for performance.
end = time.time()
print("Write Time ARROW:", end - start)

# 4. Load Data from Arrow Format to Check
start = time.time()
df2 = pd.read_feather("data_1000.arrow")   #  READ arrow file. Reads the arrow/feather file.Save to df2.
end = time.time()
print("Read Time ARROW:", end - start)

# 5. Display Arrow Data (only first 5 records)
print("Sample Data from ARROW File:")
print(df2.head())
```

#orc 1000

```python
# ---------------------------------------------
# 1. Import Libraries
# ---------------------------------------------
from faker import Faker
import pandas as pd
import pyarrow as pa
import pyarrow.orc as orc         # Special module for reading and writing ORC files.
import time

# ---------------------------------------------
# 2. Generate Fake Data (1000 Records)
# ---------------------------------------------
fake = Faker()
data = []

for _ in range(1000):
    data.append({
         "full_name": fake.name(),                                  # String
            "age": random.randint(18, 70),                             # Integer
            "salary": round(random.uniform(30000, 150000), 2),         # Float
            "is_active": random.choice([True, False]),                 # Boolean
            "register_date": fake.date(),                              # Date
            "login_time": fake.time(),                                 # Time
            "timestamp": fake.date_time(),                             # Datetime
            "gender": random.choice(["Male", "Female", "Other"]),      # Category
            "user_id": str(uuid.uuid4()),                              # UUID
            "favorite_numbers": [random.randint(1, 100) for _ in range(3)]  # List (array)
    })

df = pd.DataFrame(data)                         #Converts the list of 1000 dictionaries into a pandas table.

# 3. Convert DataFrame to PyArrow Table
table = pa.Table.from_pandas(df)                #ORC files cannot be written directly from pandas. ORC requires Arrow Table format.So we convert the pandas DataFrame Arrow Table.


# 4. Save Data in ORC Format (.orc file)
start = time.time()
with pa.OSFile("data_1000.orc", "wb") as file:      #pa.OSFile opens a file that Arrow can use."wb" = write binary mode
    orc.write_table(table, file)                    # Writes the Arrow Table into ORC format.Converts columnar Arrow ORC storage format.
end = time.time()
print("Write Time ORC:", end - start)

# ---------------------------------------------
# 5. Load ORC File to Check
# ---------------------------------------------
start = time.time()
with pa.OSFile("data_1000.orc", "rb") as file:
    orc_reader = orc.ORCFile(file)
    df2 = orc_reader.read().to_pandas()   # <-- READ ORC FILE
end = time.time()
print("Read Time ORC:", end - start)

# ---------------------------------------------
# 6. Display ORC Data (only first 5 records)
# ---------------------------------------------
print("Sample Data from ORC File:")
print(df2.head())
```
